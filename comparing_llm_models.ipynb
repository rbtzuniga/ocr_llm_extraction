```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# show all columns in pandas dataframe
pd.set_option('display.max_columns', None)
```

```python
dir = 'G:/.shortcut-targets-by-id/1a2KJDruB4ByO9oJKAZto52DF4dvwN3Y3/grain_and_milling/OCR_google_api/model_comparison'
df_haiku = pd.read_csv(f'{dir}/2013_haiku.csv')
df_sonnet = pd.read_csv(f'{dir}/2013_sonnet.csv')
display(df_haiku.head(2))
display(df_sonnet.head(2))
```

```python
def get_basic_stats(df):
    nrows = df.shape[0]
    n_unique_imgs = df['img_name'].nunique()
    # number of img_name with more than 1 rows
    n_imgs_with_multiple_rows = df.groupby('img_name').size().gt(1).sum()
    # percentiles of cap_up
    percentiles = df['cap_up'].quantile([0.25, 0.5, 0.75]).to_dict()
    # number of missing values in cap_up
    n_missing_cap_up = df['cap_up'].isna().sum()
    # percentiles of cap_fl
    percentiles_cap_fl = df['cap_fl'].quantile([0.25, 0.5, 0.75]).to_dict()
    # number of missing values in cap_fl
    n_missing_cap_fl = df['cap_fl'].isna().sum()
    # number of li == Fed
    n_li_fed = (df['li'] == 'Fed').sum()
    # number of li == St
    n_li_st = (df['li'] == 'St').sum()
    # as dataframe
    stats_df = pd.DataFrame({
        'nrows': [nrows],
        'n_unique_imgs': [n_unique_imgs],
        'n_imgs_with_multiple_rows': [n_imgs_with_multiple_rows],
        'cap_up_25th_percentile': [percentiles[0.25]],
        'cap_up_50th_percentile': [percentiles[0.5]],
        'cap_up_75th_percentile': [percentiles[0.75]],
        'n_missing_cap_up': [n_missing_cap_up],
        'cap_fl_25th_percentile': [percentiles_cap_fl[0.25]],
        'cap_fl_50th_percentile': [percentiles_cap_fl[0.5]],
        'cap_fl_75th_percentile': [percentiles_cap_fl[0.75]],
        'n_missing_cap_fl': [n_missing_cap_fl],
        'n_li_fed': [n_li_fed],
        'n_li_st': [n_li_st]
    })
    return stats_df
```

```python
df_haiku_stats = get_basic_stats(df_haiku)
df_sonnet_stats = get_basic_stats(df_sonnet)

# comparing side by side
comparison_df = pd.concat([df_haiku_stats, df_sonnet_stats], ignore_index=True)
comparison_df.index = ['Haiku', 'Sonnet']
display(comparison_df)
```

```python
# filtering haiku to img_name with a sinfle row
df_haiku_single_row = df_haiku.groupby('img_name').filter(lambda x: len(x) == 1)
display(df_haiku_single_row.head(2))
```

```python
# combining for comparison
df_comb_sonnet = df_sonnet[['img_name','elevator_type','company','city','state','cap_up','cap_fl','li']].copy()
df_comb_haiku = df_haiku_single_row[['img_name','elevator_type','company','city','state','cap_up','cap_fl','li']].copy()
df_comb = pd.merge(df_comb_haiku, df_comb_sonnet, on='img_name', suffixes=('_haiku', '_sonnet'))

cols_to_compare = ['elevator_type', 'company', 'city', 'state', 'cap_up', 'cap_fl', 'li']
# sort cols side by side for easier visual comparison
cols_sorted = ['img_name']
for col in cols_to_compare:
    cols_sorted.append(f'{col}_haiku')
    cols_sorted.append(f'{col}_sonnet')
df_comb = df_comb[cols_sorted]
display(df_comb.head(2))
```

```python

for col in cols_to_compare:
    haiku_col = f'{col}_haiku'
    sonnet_col = f'{col}_sonnet'
    df_comb[f'{col}_match'] = df_comb[haiku_col] == df_comb[sonnet_col]
    # if both are nan, consider it a match
    df_comb[f'{col}_match'] = df_comb[f'{col}_match'] | (df_comb[haiku_col].isna() & df_comb[sonnet_col].isna())

# summary of matches
match_summary = {}
for col in cols_to_compare:
    match_col = f'{col}_match'
    n_matches = df_comb[match_col].sum()
    n_total = len(df_comb)
    match_summary[col] = {
        'n_matches': n_matches,
        'n_total': n_total,
        'percent_match': n_matches / n_total * 100
    }
match_summary_df = pd.DataFrame(match_summary).T
display(match_summary_df)
```
